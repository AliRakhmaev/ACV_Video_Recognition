# Prepare the dataset according to TimesFormer format

```python
from distutils.dir_util import copy_tree

# copy subdirectory example
fromDirectory = '/content/drive/MyDrive/UCF_Crime_Dataset_Test'
toDirectory = 'merged_data'

copy_tree(fromDirectory, toDirectory)
```

```python
# copy subdirectory example
fromDirectory = '/content/drive/MyDrive/UCF_Crime_Dataset_Train'
toDirectory = 'merged_data'

copy_tree(fromDirectory, toDirectory)
```

```python
import os

# Where to save all csv files
path = "/content/merged_data"  

with open("/content/drive/MyDrive/UCF_Crime_Annotation_Train/gluon_annotations_with_normal.txt", 'r', encoding='utf-8') as anno_train:
    with open(os.path.join(path, "train.csv"), 'w+', encoding='utf-8') as output_file:
      for line in anno_train.read().splitlines():
          words = [word for word in line.split(' ') if len(word) > 0]
          if len(words) == 0:
                continue

          video_name = words[0]
          video_label = words[2]

          output_file.write(os.path.join(path, video_name) + " " + str(video_label) + "\n")
```

```python
with open("/content/drive/MyDrive/UCF_Crime_Annotation_Test/gluon_annotations.txt", 'r', encoding='utf-8') as anno_test:
  with open(os.path.join(path, "val.csv"), 'w+', encoding='utf-8') as output_file_1:
    with open(os.path.join(path, "test.csv"), 'w+', encoding='utf-8') as output_file_2:
      for line in anno_test.read().splitlines():
            words = [word for word in line.split(' ') if len(word) > 0]

            if len(words) == 0:
                continue

            video_name = words[0]
            video_label = words[2]

            output_file_1.write(os.path.join(path, video_name) + " " + str(video_label) + "\n")
            output_file_2.write(os.path.join(path, video_name) + " " + str(video_label) + "\n")
```

```python
if os.path.exists(os.path.join(path, "val.csv")):
  print("all is right")
```
