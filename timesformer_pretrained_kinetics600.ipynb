```python
import os

# Where to save all csv files
path = './merged_data' 

with open('../input/trimmed-ucf-annotations/gluon_annotations_with_normal.txt', 'r', encoding='utf-8') as anno_train:
    with open(os.path.join(path, "train.csv"), 'w+', encoding='utf-8') as output_file:
        for line in anno_train.read().splitlines():
            words = [word for word in line.split(' ') if len(word) > 0]
            if len(words) == 0:
                continue

            video_name = words[0]
            video_label = words[2]

            output_file.write(os.path.join(path, video_name) + " " + str(video_label) + "\n")
```

```python
with open("../input/trimmed-ucf-test-annotations/gluon_annotations.txt", 'r', encoding='utf-8') as anno_test:
    with open(os.path.join(path, "val.csv"), 'w+', encoding='utf-8') as output_file_1:
        with open(os.path.join(path, "test.csv"), 'w+', encoding='utf-8') as output_file_2:
            for line in anno_test.read().splitlines():
                words = [word for word in line.split(' ') if len(word) > 0]

                if len(words) == 0:
                    continue

                video_name = words[0]
                video_label = words[2]

                output_file_1.write(os.path.join(path, video_name) + " " + str(video_label) + "\n")
                output_file_2.write(os.path.join(path, video_name) + " " + str(video_label) + "\n")
```

```python
if os.path.exists(os.path.join(path, "val.csv")):
    print("all is right")
```

# Training and yaml changing

```python
# Download weights for 8 frame 224 resolution Timesformer fine-tuned on Kinetics-600
!wget https://www.dropbox.com/s/4h2qt41m2z3aqrb/TimeSformer_divST_8x32_224_K600.pyth?dl=0
```

```python
#customize iPython writefile so we can write variables
import os
from IPython.core.magic import register_line_cell_magic

if os.path.exists('./TimeSformer/configs/Kinetics/TimeSformer_divST_8x32_224.yaml'): os.remove('./TimeSformer/configs/Kinetics/TimeSformer_divST_8x32_224.yaml')

@register_line_cell_magic
def writetemplate(line, cell):
    with open(line, 'w') as f:
        f.write(cell.format(**globals()))
```

```python
%%writetemplate ./TimeSformer/configs/Kinetics/TimeSformer_divST_8x32_224.yaml

TRAIN:
  ENABLE: True
  DATASET: kinetics
  BATCH_SIZE: 7
  EVAL_PERIOD: 5
  CHECKPOINT_PERIOD: 5
  AUTO_RESUME: True
  CHECKPOINT_FILE_PATH: ./TimeSformer_divST_8x32_224_K600.pyth?dl=0
  FINETUNE: True
DATA:
  PATH_TO_DATA_DIR: ./merged_data
  NUM_FRAMES: 8
  SAMPLING_RATE: 32
  TRAIN_JITTER_SCALES: [256, 320]
  TRAIN_CROP_SIZE: 224
  TEST_CROP_SIZE: 224
  INPUT_CHANNEL_NUM: [3]
TIMESFORMER:
  ATTENTION_TYPE: 'divided_space_time'
SOLVER:
  BASE_LR: 0.001
  LR_POLICY: steps_with_relative_lrs
  STEPS: [0, 11, 14]
  LRS: [1, 0.1, 0.01]
  MAX_EPOCH: 15
  MOMENTUM: 0.9
  WEIGHT_DECAY: 1e-4
  OPTIMIZING_METHOD: sgd
MODEL:
  MODEL_NAME: vit_base_patch16_224
  NUM_CLASSES: 14
  ARCH: vit
  LOSS_FUNC: cross_entropy
  DROPOUT_RATE: 0.5
TEST:
  ENABLE: True
  DATASET: kinetics
  BATCH_SIZE: 8
  NUM_ENSEMBLE_VIEWS: 1
  NUM_SPATIAL_CROPS: 3
DATA_LOADER:
  NUM_WORKERS: 2
  PIN_MEMORY: True
NUM_GPUS: 1
NUM_SHARDS: 1
RNG_SEED: 0
OUTPUT_DIR: .
```
